```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('sales.db')
cursor = conn.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS orders (
    order_id INTEGER PRIMARY KEY,
    order_date TEXT,
    amount REAL,
    product_id INTEGER
)
''')
```

```python
cursor.execute('DELETE FROM orders')
```

```python
cursor.executemany('''
INSERT INTO orders (order_id, order_date, amount, product_id)
VALUES (?, ?, ?, ?)
''', [
    (1, '2023-01-15', 250.0, 101),
    (2, '2023-01-20', 300.0, 102),
    (3, '2023-02-05', 150.0, 103),
    (4, '2023-02-25', 500.0, 104),
    (5, '2023-03-10', 700.0, 105),
    (6, '2023-03-15', 100.0, 106),
    (7, '2024-01-10', 400.0, 107),
    (8, '2024-01-25', 350.0, 108),
    (9, '2024-02-05', 200.0, 109),
    (10, '2024-02-20', 250.0, 110)
])
```

```python
conn.commit()
```

```python
df = pd.read_sql('SELECT * FROM orders', conn)
```

```python
print("Data Preview:")
```

```python
print(df.head())
```

```python
print("\nData Info:")
print(df.info())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('month')['amount'].sum()
print("\nMonthly Sales:")
print(monthly_sales)
```

```python
monthly_sales.plot(kind='bar', title="Monthly Sales")
plt.ylabel("Sales Amount")
plt.xlabel("Month")
plt.tight_layout()
plt.show()
plt.savefig('sales_chart.png')
```

```python
import seaborn as sns
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))
monthly_sales.plot(kind='bar', ax=ax, color=sns.color_palette("coolwarm", len(monthly_sales)))
ax.set_title("Monthly Sales Overview", fontsize=18, fontweight='bold', color='navy')
ax.set_xlabel("Month", fontsize=14, color='black')
ax.set_ylabel("Sales Amount (in )", fontsize=14, color='black')
ax.tick_params(axis='both', labelsize=12, length=5)
ax.grid(True, linestyle='--', alpha=0.7)

# Annotate the bars with the sales values
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                xytext=(0, 5), textcoords='offset points',
                ha='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
plt.savefig('sales_chart.png')
```
